```python
import requests
import pandas as pd
from io import StringIO

years = [2324, 2223, 2122, 2021, 1920, 1819, 1718, 1617, 1516, 1415]
leagues = ['E0', 'E1', 'E2']

# Initialize an empty list to store dataframes
all_data = []

for year in years:
    for league in leagues:
        url = f'https://www.football-data.co.uk/mmz4281/{year}/{league}.csv'
        print(f"Fetching data from {url}")
        
        try:
            # Send a GET request to fetch the raw CSV data
            response = requests.get(url)
            response.raise_for_status()  # Check that the request was successful

            # Read the CSV data into a DataFrame
            data = pd.read_csv(StringIO(response.text))

            # Add a column to track the year and league
            data['Year'] = year
            data['League'] = league

            # Append the dataframe to the list
            all_data.append(data)

        except requests.exceptions.RequestException as e:
            print(f"Failed to fetch data for {year} {league}: {e}")

# Concatenate all dataframes into one
combined_data = pd.concat(all_data, ignore_index=True)

# Display the first few rows of the combined DataFrame
print(combined_data.head())

```

```python
date_formats = ["%d/%m/%Y", "%m/%d/%Y", "%Y-%m-%d"]

def parse_date(date_string):
    for fmt in date_formats:
        try:
            return datetime.strptime(date_string, fmt)
        except ValueError:
            continue
    return pd.NaT  # Return NaT if no format matches


df_final2['Date'] = df_final2['Date'].apply(parse_date)
```

```python
# Function to convert time strings to timedelta
def convert_to_timedelta(time_string):
    hours, minutes = map(int, time_string.split(':'))
    return pd.Timedelta(hours=hours, minutes=minutes)

# Apply the function to the 'Time' column
df_final2['Time'] = df_final2['Time'].apply(convert_to_timedelta)
```

```python
# Example dataframe creation for demonstration
import pandas as pd

df = pd.DataFrame(df_final2)

# Total goals scored by each team (Home and Away)
total_home_goals = df.groupby('HomeTeam')['FTHG'].sum().rename('TotalHomeGoals')
total_away_goals = df.groupby('AwayTeam')['FTAG'].sum().rename('TotalAwayGoals')

# Total goals conceded by each team (Home and Away)
total_home_conceded = df.groupby('HomeTeam')['FTAG'].sum().rename('TotalHomeConceded')
total_away_conceded = df.groupby('AwayTeam')['FTHG'].sum().rename('TotalAwayConceded')

# Total goals in each division
total_goals_per_div = df.groupby('Div').apply(lambda x: x['FTHG'].sum() + x['FTAG'].sum()).rename('TotalGoals')

# Number of matches played by each team
matches_played_home = df['HomeTeam'].value_counts().rename('MatchesPlayedHome')
matches_played_away = df['AwayTeam'].value_counts().rename('MatchesPlayedAway')

# Average goals scored per match
avg_home_goals = df.groupby('HomeTeam')['FTHG'].mean().rename('AvgHomeGoals')
avg_away_goals = df.groupby('AwayTeam')['FTAG'].mean().rename('AvgAwayGoals')

# Combine all results into a single dataframe for better visibility
team_stats = pd.DataFrame({
    'TotalHomeGoals': total_home_goals,
    'TotalAwayGoals': total_away_goals,
    'TotalHomeConceded': total_home_conceded,
    'TotalAwayConceded': total_away_conceded,
    'MatchesPlayedHome': matches_played_home,
    'MatchesPlayedAway': matches_played_away,
    'AvgHomeGoals': avg_home_goals,
    'AvgAwayGoals': avg_away_goals
}).fillna(0)  # Fill NaN values with 0 for teams that only played home or away

print(team_stats)
print(total_goals_per_div)
```
